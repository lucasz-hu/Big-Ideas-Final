```python
import ssl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ssl._create_default_https_context = ssl._create_unverified_context

#pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

data = pd.read_csv("BigIdeasFinalData.csv", 
                  index_col="neighborhood", 
                    header=0,
                        parse_dates=True)

data['rating'] = ((data['playgrounds'] + data['rinks'] + data['pools'])/data['children']).replace(np.inf, 0).replace(np.NaN, 0)
dataRatings = data['rating'].sort_values(ascending=False)
dataRatings.plot.barh(title="Neighborhood S rating", figsize=(15,22), fontsize=13, rot=0).invert_yaxis()
print(dataRatings)
```

As the graph shows, Regent Square comfortably takes our S-Rating top spot. Followed by Hays, then Allegheny Center. Regent Square wins by having only 52 kids. To compound this, Regent Square has 15 attractions for those 52 kids, quite a lot for the youngins to do. Hays has 21 kids with 3 attractions. Allegheny Center has 202 kids and 23 attractions. And in 20th spot is our South Oakland!
